# 基于规则, 得出test的label

```python
import pandas as pd
import numpy as np
```

```python
test = pd.read_csv('../2_data/test_sequence.csv', header=None)
test.head()
```

```python
test.shape
```

```python
test.iloc[0, 0]
```

```python
output = []
for i in range(test.shape[0]):
    a1 = test.iloc[i, 0] * test.iloc[i, 4] * test.iloc[i, 8]
    a2 = test.iloc[i, 1] * test.iloc[i, 5] * test.iloc[i, 9]
    a3 = test.iloc[i, 2] * test.iloc[i, 6] * test.iloc[i, 10]
    a4 = test.iloc[i, 3] * test.iloc[i, 7] * test.iloc[i, 11]
    
    a5 = max(a1, a2, a3, a4)
    
    if a1 == a5:
        b1 = min(test.iloc[i, 0], test.iloc[i, 4], test.iloc[i, 8])
        if b1 == test.iloc[i, 0]:
            output.append(0)
        elif b1 == test.iloc[i, 4]:
            output.append(4)
        else:
            output.append(8)
    elif a2 == a5:
        b1 = min(test.iloc[i, 1], test.iloc[i, 5], test.iloc[i, 9])
        if b1 == test.iloc[i, 1]:
            output.append(1)
        elif b1 == test.iloc[i, 5]:
            output.append(5)
        else:
            output.append(9)
    elif a3 == a5:
        b1 == min(test.iloc[i, 2], test.iloc[i, 6], test.iloc[i, 10])
        if b1 == test.iloc[i, 2]:
            output.append(2)
        elif b1 == test.iloc[i, 6]:
            output.append(6)
        else:
            output.append(10)
    else:
        b1 = min(test.iloc[i, 3], test.iloc[i, 7], test.iloc[i, 11])
        if b1 == test.iloc[i, 3]:
            output.append(3)
        elif b1 == test.iloc[i, 7]:
            output.append(7)
        else:
            output.append(11)
        
len(output)            
```

```python
output
```

```python
df = pd.DataFrame(output)
```

```python
df.to_csv('../4_result/test_label.csv', header=None, index=0)
```
